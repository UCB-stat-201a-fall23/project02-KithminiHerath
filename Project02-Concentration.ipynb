### 0. Setup 

We will consider $X_1, X_2, \ldots, X_n$ a sequence of i.i.d. random variables with mean $\mathbb{E}[X_1] = \mu$. We define the sample mean as 
$$
\bar X_n = \frac{1}{n} \sum_{i=1}^n X_i
$$
The objective of this project is to study the behavior of $\bar X_n$ as $n \rightarrow \infty$. 

### 1. Law of Large Numbers

Here we are going to focus in the convergence in probability and almost surely of the sample mean of i.i.d. random variables. Remember that a sequence of random variables $Y_1, Y_2, \ldots$ converges in probability to a random variable $Y$ if for every $\epsilon > 0$ we have 
$$
\lim_{n \rightarrow \infty} \mathbb P \left( | Y_n - Y | > \epsilon \right) = 0
$$
(the tails of the distribution goes to zero)
1) The **weak law of large numbers** states 

> The sample mean $\bar X_n$ converges in probability to $\mu$ (population mean).

2) The **strong law of large numbers** states a more strict sense in which this convergence happens

> The sample mean $\bar X_n$ converges almost surely to $\mu$. 

> $P(\bar X_n \rightarrow \mu) = 1 $

if 2) -> 1) but if 1) !-> 2)

Let's see how these two behave numerically.

#### 1.3. Compute the probability

Now, instead of evaluating the value of $\bar X_n$, compute the probability $\mathbb P \left( | \bar X_n | > \epsilon \right)$ for different values of $n$ and $\epsilon$ and plot the result as a function of $n$ (For a $X_i \sim N(0,1)$). What do you observe? See the next item for an example of the solution for this problem.

````python
np.random.seed(10)
n_sim = 800
epsilon = [0.01, 0.1, 0.2]
ns = np.arange(100,20000,100)
for eps in epsilon:
    probabilities = []
    for n in ns:
        multiple_samples = np.random.normal(loc=0.0, scale=1.0, size=(n,n_sim))
        p = (abs(np.mean(multiple_samples,axis=0)) > eps).sum()/n_sim
        probabilities.append(p)

    plt.scatter(ns, probabilities, label=r"$\epsilon = $"+f"{eps:.3f}", s=10)
    plt.plot(ns,probabilities)
    plt.xlim(0,1000)
    
plt.xlabel("n")
plt.ylabel(r"$ P \left( | \bar X_n | > \epsilon \right)$")
plt.legend()
plt.show()
````

As $\epsilon$ reduces, the probability that the mean of the sample is greater than $\epsilon$ increases for smaller sample sizes. However, the trend in the curve goes down suggesting that when the sample size increases, the weak law of large numbers has a greater chance of being true.

#### 1.4. Concentration bounds

Repeat the same plot that in the last item and show how it compares with the upper bound obtain from using Chebyshev's inequality
$$
\mathbb P \left( | \bar X_n | > \epsilon \right) \leq \frac{1}{n \epsilon^2}
$$
and the Chernoff bound
$$
\mathbb P \left( | \bar X_n | > \epsilon \right) \leq 2 \exp \left( - \frac{\epsilon^2 n}{2} \right)
$$
At the end of the day, you should obtain something like this

<img src="solution_1_34.png" alt="Solution" />

````python
np.random.seed(10)
n_sim = 800
epsilon = [0.01, 0.05, 0.08, 0.1, 0.2]
ns = np.arange(100,20000,100)
for eps in epsilon:
    probabilities = []
    chebyshev_bounds = 1.0/(ns*(eps**2))
    chernoff_bounds = 2*np.exp(-((eps**2)*ns)/2)
    for n in ns:
        multiple_samples = np.random.normal(loc=0.0, scale=1.0, size=(n,n_sim))
        p = (abs(np.mean(multiple_samples,axis=0)) > eps).sum()/n_sim
        probabilities.append(p)

    plt.figure(figsize=(10,5))
    plt.plot(ns,probabilities,marker='o',label="Simulated",markersize=5)
    plt.plot(ns,chebyshev_bounds,label="Chebyshev bound",ls='--')
    plt.plot(ns,chernoff_bounds,label="Chernoff bound",ls='--')
    if eps>=0.1:
        plt.ylim(0,0.5)
        plt.xlim(0, 1500)
    else:
        plt.ylim(0,1)
        plt.xlim(0, 20000)
    plt.xlabel("n")
    plt.title(r"$\epsilon = $"+f"{eps:.2f}")
    plt.legend()
    plt.show()
````

### 2. Central Limit theorem 

Let's first state one more time the Central Limit Theorem

> Consider $X_1, X_2, \ldots, X_n$ a sequence of i.i.d. random variables with mean $\mathbb{E}[X_1] = \mu$ and finite variance $\mathbb{V}ar[X_1] = \sigma^2$. If call 
$$
\bar X_n = \frac{1}{n} \sum_{i=1}^n X_i
$$
the sample mean, then 
$$
\sqrt{n} \frac{\bar X_n - \mu}{\sigma}
$$
converges in distribution to the standard Normal distribution $N(0,1)$.



#### 2.1. Continuous case

Pick a continuous random variable with bounded second moment and show that the central limit theorem holds numerically. Sample the sample mean $\bar X_n$ for various numbers of $n$ and show that $\sqrt{n}(\bar X_n - \mu) / \sigma$ converges to the standard normal distribution.

This is how the solution should look like for the mean of Beta distributions and a total of 10000 simulations. 

<img src="solution_2_1.png" alt="Solution" />

````python
# Beta distribution
np.random.seed(10)

a,b = 1,2
mu = float(a)/(a+b)
sd = np.sqrt((a*b)/(((a+b)**2)*(a+b+1)))
n_sim = 10000
ns = [10,100,1000]
probabilities = []

norm_pdf = stats.norm.pdf(np.linspace(-4, 4, 1000), loc=0, scale=1)

for n in ns:
    beta_multiple_samples = np.random.beta(a,b,size = (n,n_sim))
    beta_means = np.mean(beta_multiple_samples,axis=0)
    standardized_means = (np.sqrt(n)*(beta_means-mu))/sd
    plt.hist(standardized_means,bins = np.linspace(-4, 4, 100), density=True, alpha=0.7, label="sample mean")
    plt.plot(np.linspace(-4, 4, 1000), norm_pdf, label = "normal pdf")
    plt.title(f"Sample mean of Beta distribution | n = {n}")
    plt.legend(prop={'size': 13})
    plt.show()
````

#### 2.2. Discrete case

Repeat the experiment but now with a discrete random variable. When plotting the histogram, be careful on how you define the bins. 


````python
# Poisson distribution
np.random.seed(10)

lambda_=3
sd = np.sqrt(lambda_)
n_sim = 10000
ns = [10,100,1000]
probabilities = []

norm_pdf = stats.norm.pdf(np.linspace(-4, 4, 1000), loc=0, scale=1)

for n in ns:
    poisson_multiple_samples = np.random.poisson(lam=lambda_,size = (n,n_sim))
    poisson_means = np.mean(poisson_multiple_samples,axis=0)
    standardized_means = (np.sqrt(n)*(poisson_means-lambda_))/sd
    plt.hist(standardized_means,bins = np.unique(standardized_means), density=True, alpha=0.7, label="sample mean")
    plt.plot(np.linspace(-4, 4, 1000), norm_pdf, label = "normal pdf")
    plt.title(f"Sample mean of Poisson distribution | n = {n}")
    plt.legend(prop={'size': 13})
    plt.show()
````
